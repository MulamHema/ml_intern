```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
import matplotlib
```

```python
data = pd.read_csv('/content/segmentation data.csv')
```

```python
# Select two features for visualization
Feature1 = 'Age'
Feature2 = 'Income'

# Visualize the selected features
plt.scatter(data[Feature1], data[Feature2], alpha=0.5)
plt.title('Visualization of Two Features')
plt.xlabel(Feature1)
plt.ylabel(Feature2)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data[[Feature1, Feature2]])
```

```python
from sklearn.cluster import KMeans
k_means=KMeans(n_clusters=4,random_state=42)
k_means.fit(data[[Feature1,Feature2]])
```

```python
data['KMeans_labels']=k_means.labels_

# Plotting resulting clusters
colors=['purple','red','orange','yellow']
plt.figure(figsize=(6,6))
plt.scatter(data[Feature1],data[Feature2],c=data['KMeans_labels'],cmap=matplotlib.colors.ListedColormap(colors),s=5)
plt.title('K-Means Clustering',fontsize=20)
plt.xlabel('Feature1',fontsize=14)
plt.ylabel('Feature2',fontsize=14)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
model = AgglomerativeClustering(n_clusters=4, affinity='euclidean')
model.fit(data[[Feature1,Feature2]])
```

```python
data['HR_labels']=model.labels_

# Plotting resulting clusters
plt.figure(figsize=(6,6))
plt.scatter(data[Feature1],data[Feature2],c=data['HR_labels'],cmap=matplotlib.colors.ListedColormap(colors),s=5)
plt.title('Hierarchical Clustering',fontsize=20)
plt.xlabel('Feature1',fontsize=14)
plt.ylabel('Feature2',fontsize=14)
plt.show()
```

```python
from sklearn.cluster import DBSCAN
import pandas as pd


# Initialize DBSCAN
dbscan = DBSCAN()

# Fit DBSCAN to the data
dbscan.fit(data[[Feature1, Feature2]])

# Accessing the cluster labels assigned by DBSCAN
cluster_labels = dbscan.labels_
```

```python
# Plotting resulting clusters
plt.figure(figsize=(6,6))
plt.scatter(data[Feature1],data[Feature2],c=cluster_labels,cmap=matplotlib.colors.ListedColormap(colors),s=5)
plt.title('DBSCAN Clustering',fontsize=20)
plt.xlabel('Feature1',fontsize=14)
plt.ylabel('Feature2',fontsize=14)
plt.show()
```

```python
from sklearn.neighbors import NearestNeighbors
neigh = NearestNeighbors(n_neighbors=2)
nbrs = neigh.fit(data[[Feature1,Feature2]])
distances, indices = nbrs.kneighbors(data[[Feature1,Feature2]])
```

```python
# Plotting K-distance Graph
distances = np.sort(distances, axis=0)
distances = distances[:,1]
plt.figure(figsize=(10,5))
plt.plot(distances)
plt.title('K-distance Graph',fontsize=20)
plt.xlabel('Data Points sorted by distance',fontsize=14)
plt.ylabel('Epsilon',fontsize=14)
plt.show()
```

```python
from sklearn.cluster import DBSCAN
dbscan_opt=DBSCAN(eps=30,min_samples=6)
dbscan_opt.fit(data[[Feature1,Feature2]])
cluster_labels=dbscan_opt.labels_
```

```python
# Plotting the resulting clusters
plt.figure(figsize=(6,6))
plt.scatter(data[Feature1],data[Feature2],c=cluster_labels,cmap=matplotlib.colors.ListedColormap(colors),s=15)
plt.title('DBSCAN Clustering',fontsize=20)
plt.xlabel('Feature1',fontsize=14)
plt.ylabel('Feature2',fontsize=14)
plt.show()
```
